```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_excel(r"D:\CSV Files\Sales Data.xlsx")
data
```

```python
get_duplicate = data.duplicated()
empty_value = data.isna().sum()
data_fill = data.fillna(0,inplace=True)
data
```

```python
# create new month column
data["Order_date"] = pd.to_datetime(data["Order_date"]).astype("datetime64[ns]")
data["month_name"] = data["Order_date"].dt.month_name()
data
```

```python
# Change the data type in order_date
data["Order_date"] = pd.to_datetime(data["Order_date"]).astype("date32[pyarrow]")
data
```

```python
# Change column name
rename_column = data.rename(columns={"Unit_Price_USD":"Unit_Price"},inplace=True)
data
```

```python
# Saparat location
data[["State","Country"]] = data["Location"].str.split(",",expand=True)
data
```

```python
data = data.drop(columns=["Location"])
data
```

```python
total_saleing = data["Total_Price_USD"].sum()
total_saleing
```

```python
filter_states = data[data["Country"] == "USA"]
filter_states
```

```python
plt.figure(figsize=(15,5))
sales_usa_states = data.groupby("State")["Total_Price_USD"].sum().sort_values(ascending=False)
bars = plt.bar(sales_usa_states.index,sales_usa_states.values,color="green",edgecolor="black")
plt.xlabel("States")
plt.ylabel("Total sales")
plt.title("USA All states total sales")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:,.0f}',
        ha='center',
        va='bottom'
    )

plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
total_country_sales = data.groupby("Country")["Total_Price_USD"].sum().sort_values(ascending=True)
plt.figure(figsize=(15,5))
plt.plot(total_country_sales.index,total_country_sales.values,marker='o',color='red')
plt.xlabel("Country Name")
plt.ylabel("Total saleing")

for x,y in zip(total_country_sales.index,total_country_sales.values):
    plt.text(
        x,
        y,
        f'{y:,.0f}',
        ha='center',
        va='bottom'
    )
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.grid()
plt.show()
```
